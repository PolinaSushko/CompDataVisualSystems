```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Прочитайте CSV файл (використовуйте метод read_csv)
df = pd.read_csv("Global YouTube Statistics.csv", encoding="latin1")
df
```

```python
# 3. Виведіть розміри датасету (використовуйте метод shape)
df.shape
```

```python
# 4. Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків  у кожному зі стовпців (використовуйте методи isna та sum).
missing_data = df.isna().sum()
print("Кількість пропусків у кожному стовпці: ")
print(missing_data)
```

```python
# Якщо є пропуски, замінить їх на NaN, потім тип даних на числовий.

# При завантажені цієї таблиці пропущені данні одразу були замінені на NaN.
# Типи даних теж не потрібно змінювати, оскільки стовпці із числовими даними мають типи int64 або float64, а стовпці із текстовими даними -object. 
df.dtypes
```

```python
# 5. Замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float.
numerical_cols = df.select_dtypes(['float64', 'int64']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
df
```

```python
df = df.dropna().reset_index(drop=True)
df
```

```python
# 6. Ще раз перевірте кількість пропусків.
df.isna().sum()
```

```python
# 7. Перевірте, скільки унікальних країн представлено у колонці "Country"
unique_countries = df['Country'].unique()
print(f"У колонці 'Country' представлено {len(unique_countries)} унікальних країн.")
```

```python
# 8. Побудуйте діаграму розподілу переглядів (використовуйте kind='hist')
df['video views'].plot(kind='hist', title='Розподіл переглядів')
```

```python
# 9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)
print(f"Максимальна кількість переглядів: {df['video views'].max()}")
print(f"Мінімальна кількість переглядів: {df['video views'].min()}")
print(f"Середня кількість переглядів: {df['video views'].mean()}")
```

```python
# 10. Знайдіть країну, де найбільше відео було завантажено на YouTube
most_videos_country = df['Country'].value_counts().idxmax()
print(f"Країна з найбільшою кількістю завантажених відео на YouTube: {most_videos_country}")
```

```python
# 11. Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads.
videos_most_views = df.iloc[df['uploads'].idxmax()]
videos_least_views = df.iloc[df['uploads'].idxmin()]

print(f"Відео з найбільшою кількістю завантажень: {videos_most_views['Title']}")
print(f"Відео з найменшою кількістю завантажень: {videos_least_views['Title']}")
```
