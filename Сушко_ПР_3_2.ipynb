## Завдання 2

Побудуйте графік поверхні (3D surface plot) для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn.
z=sin⁡√(x^2+y^2), x∈[-5;5],y∈[-5;5]

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(13, 6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='inferno')

ax.set_title('Графік поверхні: z=sin⁡√(x^2+y^2), x∈[-5;5],y∈[-5;5]', loc='center')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-1, 1)

plt.colorbar(surface, ax=ax)

plt.show()
```
